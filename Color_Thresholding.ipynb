```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image= cv2.imread("/content/giraffe-Kenya.png")
```

```python
rgb_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
hsv_img=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(rgb_img)
```

```python
lower_white=np.array([0,0,190])
upper_white=np.array([30,80,255])
```

```python
mask=cv2.inRange(hsv_img,lower_white,upper_white)
```

```python
plt.imshow(mask)
```

```python
black_hsv=np.uint8([[[120,255,0]]])
black_bgr=cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)
```

```python
image_result=image.copy()
```

```python
image_result[mask>0]=black_bgr
```

```python
image_result_rgb=cv2.cvtColor(image_result,cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,3,1)
plt.imshow(rgb_img)
plt.subplot(1,3,2)
plt.imshow(mask)
plt.subplot(1,3,3)
plt.imshow(image_result_rgb)
```
