```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("/content/giraffe-Kenya.png")
```

```python
rgb_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

###1. Image Resizing

```python
rgb_img.shape
```

```python
new_height=400
new_width=400

resized_img=cv2.resize(rgb_img,(new_width,new_height))
plt.imshow(resized_img)
```

###2.Image Scaling

```python
scale_factor=50
s_width=int(rgb_img.shape[1]*scale_factor/100)
s_hight=int(rgb_img.shape[0]*scale_factor/100)

scaled_image=cv2.resize(rgb_img,(s_width,s_hight), interpolation=cv2.INTER_AREA)
plt.imshow(scaled_image)
print(scaled_image.shape)
```

###3. INTERPOLATION

```python
resize_width=600
resize_hight=600

img=cv2.resize(rgb_img,(resize_width,resize_hight),interpolation=cv2.INTER_NEAREST)
plt.imshow(img)
```

```python
resize_width=600
resize_hight=600

img=cv2.resize(rgb_img,(resize_width,resize_hight),interpolation=cv2.INTER_LINEAR)
plt.imshow(img)
```

```python
resize_width=600
resize_hight=600

img=cv2.resize(rgb_img,(resize_width,resize_hight),interpolation=cv2.INTER_CUBIC)
plt.imshow(img)
```

```python
resize_width=600
resize_hight=600

img=cv2.resize(rgb_img,(resize_width,resize_hight),interpolation=cv2.INTER_LANCZOS4)
plt.imshow(img)
```
