```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("/content/giraffe-Kenya.png")
```

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

## 1. FLIP

```python
#1----> horizontally flip
#0-----> vertically flip
#-1----> Both axes

flip_horizontal=cv2.flip(rgb_img,1)
flip_vertical=cv2.flip(rgb_img,0)
flip_both=cv2.flip(rgb_img,-1)

titles=["Original","Horizontal","Vertical","Both"]
images=[rgb_img,flip_horizontal,flip_vertical,flip_both]

plt.figure(figsize=(12,8))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])
  plt.axis("off")
plt.tight_layout()
plt.show()
```

###2. ROTATE

```python
rotate_90_clockwise=cv2.rotate(rgb_img,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise=cv2.rotate(rgb_img,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180=cv2.rotate(rgb_img,cv2.ROTATE_180)

titles=["Original","90 clockwise","90 counterclockwise","180"]
images=[rgb_img,rotate_90_clockwise,rotate_90_counterclockwise,rotate_180]
plt.figure(figsize=(12,8))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i])
  plt.title(titles[i])
  plt.axis("off")
plt.tight_layout()
plt.show()
```
